```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

facebook = pd.read_csv('DATA/dataset_Facebook.csv', delimiter=";")
```

```python
ax = sns.scatterplot(x = 'comment', y= 'share', data = facebook, hue = 'Paid', ci = False)
```

```python
ax = sns.barplot(x = 'like', y = 'share', data = facebook, ci = None, hue = 'Paid')
```

```python
numeric_vars = ['like','share','comment','Total Interactions']
```

```python
from pandas.plotting import scatter_matrix
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax = scatter_matrix(facebook[numeric_vars], ax=ax, diagonal='kde')
```

```python
sct = facebook[numeric_vars]
sns.scatterplot(data = sct)
```

```python
sns.pairplot(sct)
```

```python
sns.catplot(x= 'Paid', y = 'share', data = facebook, col = 'Type', height=4, aspect=.8)
sns.catplot(x= 'Paid', y = 'comment', data = facebook, col = 'Type', height=4, aspect=.8)
sns.catplot(x= 'Paid', y = 'like', data = facebook, col = 'Type', height=4, aspect=.8)
sns.catplot(x= 'Paid', y = 'Total Interactions', data = facebook, col = 'Type', height=4, aspect=.8)

```

```python
facebook.set_index(["Type", "Paid"]).count(level="Type")
```

```python
sns.set_context("notebook")
sns.set_style("dark") 
ax = sns.lineplot(x="Post Month", y="Total Interactions", hue= 'Type', ci=None, data=facebook)
```
