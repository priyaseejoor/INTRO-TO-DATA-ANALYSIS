```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
univ = pd.read_csv('/Users/priya/Downloads/University.csv')
```

```python
df = pd.DataFrame(univ)
```

```python
univ.dtypes
```

```python
univ.columns = univ.columns.str.replace(' ', '_')
```

```python
univ.columns
```

```python
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace('$', '')
univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace('$', '')
```

```python
univ['Full_Time_Enrollment'] = univ['Full_Time_Enrollment'].str.replace(',', '')
df['Full_Time_Enrollment'] = pd.to_numeric(df['Full_Time_Enrollment'])
```

```python
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace(',', '')
df['Average_Undergraduate_Tuition_and_Fees'] = pd.to_numeric(df['Average_Undergraduate_Tuition_and_Fees'])
```

```python
univ['Percentage_of_International_Students'] = univ['Percentage_of_International_Students'].str.replace('%', '')
df['Percentage_of_International_Students'] = pd.to_numeric(df['Percentage_of_International_Students'])
```

```python
df['Percentage_of_International_Students'] = df['Percentage_of_International_Students'].div(100)
```

```python
univ['Undergraduate_Admit_Rate'] = univ['Undergraduate_Admit_Rate'].str.replace('%', '')
df['Undergraduate_Admit_Rate'] = pd.to_numeric(df['Undergraduate_Admit_Rate'])
df['Undergraduate_Admit_Rate'] = df['Undergraduate_Admit_Rate'].div(100)
```

```python
univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace(',', '')
df['Endowment_per_Full_Time_Student'] = pd.to_numeric(df['Endowment_per_Full_Time_Student'])
```
