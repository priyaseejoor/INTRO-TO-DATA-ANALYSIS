```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
univ = pd.read_csv('/Users/priya/Downloads/University.csv')
```

```python
df = pd.DataFrame(univ)
```

```python
univ.dtypes
```

```python
univ.columns = univ.columns.str.replace(' ', '_')
```

```python
univ.columns
```

```python
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace('$', '')
univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace('$', '')
```

```python
univ['Full_Time_Enrollment'] = univ['Full_Time_Enrollment'].str.replace(',', '')
df['Full_Time_Enrollment'] = pd.to_numeric(df['Full_Time_Enrollment'])
```

```python
univ['Average_Undergraduate_Tuition_and_Fees'] = univ['Average_Undergraduate_Tuition_and_Fees'].str.replace(',', '')
df['Average_Undergraduate_Tuition_and_Fees'] = pd.to_numeric(df['Average_Undergraduate_Tuition_and_Fees'])
```

```python
univ['Percentage_of_International_Students'] = univ['Percentage_of_International_Students'].str.replace('%', '')
df['Percentage_of_International_Students'] = pd.to_numeric(df['Percentage_of_International_Students'])
```

```python
df['Percentage_of_International_Students'] = df['Percentage_of_International_Students'].div(100)
```

```python
univ['Undergraduate_Admit_Rate'] = univ['Undergraduate_Admit_Rate'].str.replace('%', '')
df['Undergraduate_Admit_Rate'] = pd.to_numeric(df['Undergraduate_Admit_Rate'])
df['Undergraduate_Admit_Rate'] = df['Undergraduate_Admit_Rate'].div(100)
```

```python
univ['Endowment_per_Full_Time_Student'] = univ['Endowment_per_Full_Time_Student'].str.replace(',', '')
df['Endowment_per_Full_Time_Student'] = pd.to_numeric(df['Endowment_per_Full_Time_Student'])
```

```python
univ.dtypes
```

```python
univ.head(3)
```

```python
univ["Prof_G_Categorization"]= univ["Prof_G_Categorization"].replace('Perish', 0) 
```

```python
univ["Prof_G_Categorization"]= univ["Prof_G_Categorization"].replace('Struggle', 1) 
```

```python
univ["Prof_G_Categorization"]= univ["Prof_G_Categorization"].replace('Survive', 2) 
```

```python
univ["Prof_G_Categorization"]= univ["Prof_G_Categorization"].replace('Thrive', 3) 
```

```python
univ.head(3)
```

```python
df['Prof_G_Categorization'] = pd.to_numeric(df['Prof_G_Categorization'])
```

```python
ax =sns.scatterplot(y = 'Eduation_Score' , x = 'Rank', data = univ, hue = 'Prof_G_Categorization', palette = ['blue', 'orange', 'green', 'red'])

#I wanted to look into the categorization of whether a college/university would
#be able to surivie covid. The number 0 starts off lowest on the list meaning its predicted to become a school
#that perishes, all the way up to 3 which represents a school thats thriving. 
#As you can see from this scatterplot number 3 is mostly represented in the upper left corner.
#There is a correlation between education score and the top tier uni's, I wouldn't say its a strong 
#correlation but even looking at the dataset ordering by rank the top 20 schools are set to survive and thrive.

```

```python
agg = univ.groupby('Prof_G_Categorization')[['Endowment_per_Full_Time_Student']].max()
agg
```

```python
fts = [653798,730888,1186185,3077087]
#y = [0,1,2,3]


my_labels = 'Perish','Struggle','Survive', 'Thrive'
my_colors = ['blue','orange','green','red']
plt.pie(fts,labels=my_labels,autopct='%1.1f%%', explode = [0,0,0,0.1], startangle = 90)
plt.title('Max Endowment Per Categorization')
plt.axis('equal')
plt.show()
```
